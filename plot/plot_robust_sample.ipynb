```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_context('talk')
```

```python
sns.palplot(sns.color_palette("gray", 100))
```

```python
new_gray=sns.color_palette("gray",4)
```

```python
new_gray=[(0, 0, 0), (0.85, 0.85, 0.85)]
```

## Brazil

```python
plot_bra2 = pd.read_csv('sensi_withhold_bra.csv')
```

```python
eff_new = pd.DataFrame(
    np.array([np.repeat(list(plot_bra2['intervention']),320), 
              plot_bra2[plot_bra2.columns[2:]].values.reshape(1,-1)[0]]).T,
    columns=['intervention','x'])
eff_new['x'] = eff_new['x'].astype(float)
```

```python
eff_new['color'] =([1]*319+[0.1])*10
```

```python
fig1,ax = plt.subplots(figsize=(10,6))
ax.spines['left'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
plt.axvline(x=0,ls="-",linewidth=1,c="black")
sns.scatterplot(data = eff_new, 
                x='x',
                y='intervention',
                hue='color',
                s=200,
                palette=new_gray,  
                alpha=0.3,      
                legend=False,
                edgecolor=None,  
                ax=ax)
plt.xlabel('Brazil',c="black",fontsize=24,fontname='Helvetica') 
plt.ylabel('') 
#plt.xlim(-1.5,1)
```

```python
fig1.savefig("sensi_withhold_bra",bbox_inches='tight',dpi=300)
```
