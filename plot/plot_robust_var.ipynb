## Brazil

```python
new1 = pd.read_csv("sensi_var_bra1.csv")
```

```python
new2 = pd.read_csv("sensi_var_bra2.csv")
```

```python
fig1, ax = plt.subplots(figsize=[10,18])

ax.spines['left'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.axvline(x=0,ls="-",linewidth=0.6,c="black")

ax.tick_params(labelsize=15)


ax.errorbar(x=new1["b"][::-1], 
            y=new1["intervention"][::-1], 
            xerr=new1["se"][::-1]*1.96, 
            fmt="o", 
            markersize=6,
            capsize=2, 
            elinewidth=1, 
            barsabove=True, 
            color='dimgray')

ax.errorbar(x=new2["b"][::-1], 
            y=new2["intervention"][::-1], 
            xerr=new2["se"][::-1]*1.96, 
            fmt="o", 
            markersize=5,
            capsize=2, 
            elinewidth=1, 
            alpha=0.5,
            barsabove=True, 
            color='gray')

ax.set_yticks([9,25,41,57,73,89,105,121,137,153])
plt.xticks(size=25)
plt.yticks(size=25)
plt.xlabel('Brazil',fontsize=40,fontname='Helvetica')
plt.xlim(-1.5,1)
#plt.ylim(-4,23)
```

```python
fig1.savefig("sensi_var_bra",bbox_inches='tight',dpi=300)
```
